**Project 3 Guess number game**

```python
import random
import time
from colorama import Fore, Style

def guess_the_number():
    print(Fore.YELLOW + "ğŸ¯ Welcome to the 'Guess the Number' Game! ğŸ¤©ğŸ®" + Style.RESET_ALL)
    print(Fore.GREEN + "ğŸ¤– The computer has chosen a secret number. Can you guess it? ğŸ”¢ğŸ¤”\n" + Style.RESET_ALL)

    while True:
        # Difficulty selection
        print(Fore.CYAN + "Choose your difficulty level:" + Style.RESET_ALL)
        print("1ï¸âƒ£ Easy (1-10, Unlimited Attempts)")
        print("2ï¸âƒ£ Medium (1-20, 5 Attempts)")
        print("3ï¸âƒ£ Hard (1-50, 7 Attempts)")

        difficulty = input(Fore.MAGENTA + "Enter 1, 2, or 3: " + Style.RESET_ALL)

        if difficulty == "1":
            max_number = 10
            max_attempts = None  # Unlimited attempts
        elif difficulty == "2":
            max_number = 20
            max_attempts = 5
        elif difficulty == "3":
            max_number = 50
            max_attempts = 7
        else:
            print(Fore.RED + "âš ï¸ Invalid choice! Please enter 1, 2, or 3." + Style.RESET_ALL)
            continue

        # Generate random secret number
        secret_number = random.randint(1, max_number)
        attempts = 0  # Track attempts

        print(Fore.BLUE + f"\nğŸ” The secret number is between **1 and {max_number}**. Start guessing!" + Style.RESET_ALL)

        while True:
            try:
                user_guess = int(input(Fore.YELLOW + "ğŸ’¡ Enter your guess: " + Style.RESET_ALL))
                attempts += 1

                # Check the guess
                if user_guess < 1 or user_guess > max_number:
                    print(Fore.RED + f"âš ï¸ Please guess within the range **1 to {max_number}**!" + Style.RESET_ALL)
                    continue

                if user_guess < secret_number:
                    hint = "ğŸ“‰ Too low! " + ("You're getting closer! ğŸ”¥" if secret_number - user_guess <= 3 else "Try a higher number! â¬†ï¸")
                    print(Fore.RED + hint + Style.RESET_ALL)
                elif user_guess > secret_number:
                    hint = "ğŸ“ˆ Too high! " + ("Almost there! ğŸŒŸ" if user_guess - secret_number <= 3 else "Try a lower number! â¬‡ï¸")
                    print(Fore.RED + hint + Style.RESET_ALL)
                else:
                    print(Fore.GREEN + f"ğŸ‰ **Congratulations!** ğŸŠ You guessed the right number in **{attempts}** attempts! ğŸ†ğŸ¥³" + Style.RESET_ALL)
                    break

                # Check attempt limit
                if max_attempts and attempts >= max_attempts:
                    print(Fore.RED + f"âŒ You've used all {max_attempts} attempts! The number was {secret_number}. Better luck next time! ğŸ˜¢" + Style.RESET_ALL)
                    break

            except ValueError:
                print(Fore.RED + "âš ï¸ **Oops!** Please enter a valid **number**!" + Style.RESET_ALL)

        # Play again option
        play_again = input(Fore.CYAN + "\nğŸ”„ Do you want to play again? (yes/no): " + Style.RESET_ALL).strip().lower()
        if play_again != 'yes':
            print(Fore.YELLOW + "ğŸ® Thanks for playing! See you next time! ğŸ‘‹ğŸ˜Š" + Style.RESET_ALL)
            break

# Run the game
guess_the_number()

```

**Project 4 Rock,Paper scissor Game**

```python
import random

def play_game():
    print("\nğŸ® Welcome to Rock, Paper, Scissors! âœ‚ï¸ğŸª¨ğŸ“„")
    print("ğŸ”¹ Type 'rock', 'paper', or 'scissors' to play.")
    print("ğŸ”¹ Type 'exit' anytime to quit the game.\n")

    choices = ["rock", "paper", "scissors"]

    while True:
        user_choice = input("ğŸ¤” Choose: Rock, Paper, or Scissors? ").strip().lower()

        if user_choice == "exit":
            print("ğŸ‘‹ Thanks for playing! See you next time. ğŸ­")
            break

        if user_choice not in choices:
            print("âš ï¸ Invalid choice! Please choose Rock, Paper, or Scissors. âŒ\n")
            continue

        computer_choice = random.choice(choices)

        print(f"\nğŸ§‘ You chose: {user_choice.capitalize()} ğŸ­")
        print(f"ğŸ¤– Computer chose: {computer_choice.capitalize()} ğŸ’»")

        if user_choice == computer_choice:
            print("ğŸ˜² It's a tie! Try again. ğŸ”„\n")
        elif (user_choice == "rock" and computer_choice == "scissors") or \
             (user_choice == "paper" and computer_choice == "rock") or \
             (user_choice == "scissors" and computer_choice == "paper"):
            print("ğŸ‰ You win! ğŸ†ğŸ”¥\n")
        else:
            print("ğŸ˜¢ You lose! Better luck next time. ğŸ’€\n")

# Start the game
play_game()
```

**Project 5 Hangman**

```python
import random

# Simple list of words
words = ["python", "hangman", "laptop", "coding", "developer"]

# Select a random word
word = random.choice(words)
word_letters = set(word)  # Unique letters in the word
guessed_letters = set()  # Correct guesses
wrong_letters = set()  # Incorrect guesses
attempts = 6  # Number of chances

# Simple Hangman stages
hangman_stages = [
    "ğŸ˜µ | You lost!",
    "ğŸ˜¨ | Only 1 try left!",
    "ğŸ˜° | 2 tries left!",
    "ğŸ¥¶ | 3 tries left!",
    "ğŸ«£ | 4 tries left!",
    "ğŸ™‚ | 5 tries left!",
]

print("\nğŸ­ Welcome to Hangman! Try to guess the word!")

while attempts > 0 and word_letters:
    # Show word progress
    display_word = [letter if letter in guessed_letters else "_" for letter in word]
    print("\nWord: " + " ".join(display_word))

    # Show incorrect guesses
    if wrong_letters:
        print("âŒ Wrong letters: " + ", ".join(wrong_letters))

    # Take user input
    guess = input("ğŸ”¤ Guess a letter: ").lower()

    if len(guess) != 1 or not guess.isalpha():
        print("âš ï¸ Enter a **single valid letter**.")
        continue

    if guess in guessed_letters or guess in wrong_letters:
        print("âš ï¸ You already guessed that letter!")
        continue

    # Check the guess
    if guess in word_letters:
        guessed_letters.add(guess)
        word_letters.remove(guess)
        print("âœ… Correct!")
    else:
        attempts -= 1
        wrong_letters.add(guess)
        print(f"âŒ Wrong! {attempts} attempts left.")
        print(hangman_stages[5 - attempts])  # Show Hangman stage

# Game Over Check
if not word_letters:
    print(f"\nğŸ‰ You won! The word was: {word.upper()} ğŸ†")
else:
    print(f"\nğŸ’€ Game Over! The word was: {word.upper()}")
```

**Project 7 Password generator**


```python
import random
import string

def generate_password(length):
    # Ensuring at least one uppercase, one lowercase, one digit, and one special character
    if length < 4:
        return "âš ï¸ Password length should be at least 4 characters."

    all_chars = string.ascii_letters + string.digits + string.punctuation
    password = [
        random.choice(string.ascii_uppercase),  # At least one uppercase
        random.choice(string.ascii_lowercase),  # At least one lowercase
        random.choice(string.digits),           # At least one digit
        random.choice(string.punctuation)       # At least one special character
    ]

    # Fill the remaining length with random characters
    password += random.choices(all_chars, k=length - 4)

    # Shuffle to mix mandatory characters randomly
    random.shuffle(password)

    return "".join(password)

# Get valid input from user
while True:
    try:
        length = int(input("ğŸ” Enter password length: "))
        if length < 4:
            print("âš ï¸ Password length must be at least 4 characters.")
            continue
        break
    except ValueError:
        print("âš ï¸ Please enter a valid number.")

# Generate and display password
print("\nğŸ‰ Generated Secure Password:", generate_password(length))

```

```python
import time

def calculate_bmi(weight, height):
    return weight / (height ** 2)

def bmi_category(bmi):
    if bmi < 18.5:
        return "Underweight ğŸ˜” Eat nutritious foods! ğŸğŸ¥¦"
    elif 18.5 <= bmi < 24.9:
        return "Normal weight âœ… Great job! Keep staying healthy! ğŸ’ªâœ¨"
    elif 25 <= bmi < 29.9:
        return "Overweight âš ï¸ Time to stay active! ğŸƒâ€â™‚ï¸ğŸ¥—"
    else:
        return "Obese â— Consider a healthier lifestyle. ğŸ‹ï¸â€â™€ï¸ğŸ"

def main():
    print("ğŸ’  BMI Calculator âš–ï¸")

    try:
        weight = float(input("âš–ï¸ Enter your weight (kg): "))
        height = float(input("ğŸ“ Enter your height (m): "))

        if weight > 0 and height > 0:
            print("ğŸ”„ Analyzing your stats... ğŸ§®")
            time.sleep(1)
            bmi = calculate_bmi(weight, height)
            category = bmi_category(bmi)

            print(f"\nğŸ“Š Your BMI: {bmi:.2f}")
            print(f"ğŸ† Category: {category}")
            print("ğŸ’¡ Stay fit and eat healthy! ğŸŒ¿ğŸ’–")
        else:
            print("ğŸš¨ Please enter valid values! ğŸš¨")
    except ValueError:
        print("ğŸš¨ Invalid input! Please enter numeric values. ğŸš¨")

if __name__ == "__main__":
    main()
```

```python

import streamlit as st
import pandas as pd
import os
import random
import json

# Constants
POSTS_FILE = "posts.json"
AVATAR_URLS = [
    "https://i.pravatar.cc/40?img=1", "https://i.pravatar.cc/40?img=2",
    "https://i.pravatar.cc/40?img=3", "https://i.pravatar.cc/40?img=4"
]

# Load existing posts
def load_posts():
    if os.path.exists(POSTS_FILE):
        with open(POSTS_FILE, "r") as f:
            return json.load(f)
    return []

# Save posts
def save_posts(posts):
    with open(POSTS_FILE, "w") as f:
        json.dump(posts, f, indent=4)

# Create post
def create_post(title, content, image_url):
    posts = load_posts()
    posts.append({
        "id": len(posts) + 1,
        "title": title,
        "content": content,
        "image": image_url,
        "likes": 0,
        "comments": []
    })
    save_posts(posts)
    st.success("âœ… Post Created Successfully!")

# Add comment
def add_comment(post_id, comment):
    posts = load_posts()
    for post in posts:
        if post["id"] == post_id:
            post["comments"].append({
                "text": comment,
                "avatar": random.choice(AVATAR_URLS)  # Random avatar for user
            })
            break
    save_posts(posts)
    st.success("ğŸ’¬ Comment Added!")

# Like a post
def like_post(post_id):
    posts = load_posts()
    for post in posts:
        if post["id"] == post_id:
            post["likes"] += 1
            break
    save_posts(posts)

# Custom Styles
st.markdown("""
    <style>
        .stApp { background-color: #121212; color: #E0E0E0; }
        .title { text-align: center; font-size: 2.5rem; color: #00BFFF; text-shadow: 2px 2px 4px rgba(0, 191, 255, 0.5); }
        .card { background-color: #1E1E1E; padding: 20px; border-radius: 12px; margin-bottom: 15px; box-shadow: 0px 4px 10px rgba(255, 64, 129, 0.2); }
        .comment { background-color: #222; padding: 10px; border-radius: 8px; margin: 8px 0; }
        .comment-avatar { border-radius: 50%; margin-right: 10px; }
        .comment-text { display: flex; align-items: center; }
        .stButton > button { background-color: #00BFFF !important; color: white; border-radius: 8px; padding: 8px 15px; font-weight: bold; }
        .stButton > button:hover { background-color: #FF4081 !important; }
    </style>
""", unsafe_allow_html=True)

# UI
st.markdown("<h1 class='title'>ğŸ“ Social Post & Review App</h1>", unsafe_allow_html=True)
tabs = st.tabs(["ğŸ“Œ Create Post", "ğŸ“œ View Posts"])

with tabs[0]:  # Create Post
    st.subheader("ğŸ“ Create a New Post")
    title = st.text_input("Post Title")
    content = st.text_area("Post Content")
    image = st.text_input("Post Image URL (Optional)")
    if st.button("Publish Post"):
        if title and content:
            create_post(title, content, image)
        else:
            st.warning("âš ï¸ Please enter title and content.")

with tabs[1]:  # View Posts
    st.subheader("ğŸ“œ All Posts")
    posts = load_posts()

    if posts:
        for post in posts[::-1]:  # Show latest posts first
            st.markdown(f"""
                <div class="card">
                    <h3 style="color:#00BFFF;">ğŸ“Œ {post['title']}</h3>
                    <p>{post['content']}</p>
                    {"<img src='" + post['image'] + "' width='100%' style='border-radius: 10px;'/>" if post['image'] else ""}
                    <p style="color: #FF4081;">â¤ï¸ {post['likes']} Likes</p>
                </div>
            """, unsafe_allow_html=True)

            # Like Button
            if st.button(f"ğŸ‘ Like {post['title']}", key=f"like_{post['id']}"):
                like_post(post["id"])
                st.rerun()

            # Display Comments
            st.markdown("ğŸ’¬ **Comments:**")
            for comment in post["comments"]:
                st.markdown(f"""
                    <div class="comment">
                        <div class="comment-text">
                            <img class="comment-avatar" src="{comment['avatar']}" width="30"/>
                            {comment['text']}
                        </div>
                    </div>
                """, unsafe_allow_html=True)

            # Add Comment
            new_comment = st.text_input(f"ğŸ’­ Add Comment on '{post['title']}'", key=f"comment_{post['id']}")
            if st.button(f"Submit Comment on '{post['title']}'", key=f"submit_comment_{post['id']}"):
                if new_comment:
                    add_comment(post["id"], new_comment)
                    st.rerun()
                else:
                    st.warning("âš ï¸ Please enter a comment.")
    else:
        st.write("No posts available. Create one first!")

st.success("âœ… App Loaded Successfully!")
```
